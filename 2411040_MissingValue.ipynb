Import library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_excel('dataset_skincare.xlsx')

dataset.head()
```

cek data missing value

```python
print(dataset)

print("\nJumlah Missing Value:")
print(dataset.isnull().sum())
```

pisahkan x dan y

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

print(X)
print(y)
```

imputasi missing value

```python
from sklearn.impute import SimpleImputer

# Brand
imputer_brand = SimpleImputer(strategy='most_frequent')
X[:,1:2] = imputer_brand.fit_transform(X[:,1:2])

# Harga
imputer_harga = SimpleImputer(strategy='mean')
X[:,2:3] = imputer_harga.fit_transform(X[:,2:3])

# Stok
y = y.reshape(-1,1)

imputer_stok = SimpleImputer(strategy='mean')
y = imputer_stok.fit_transform(y)

y = y.ravel()

print(X)
print(y)
```

encoding

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(
    transformers=[
        ('encoder', OneHotEncoder(), [0,1,3])
    ],
    remainder='passthrough'
)

X = np.array(ct.fit_transform(X))

print(X)
```

split data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=1
)

print(X_train)
print(X_test)
```

feature scaling

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train[:, -1:] = sc.fit_transform(X_train[:, -1:])
X_test[:, -1:] = sc.transform(X_test[:, -1:])

print(X_train)
print(X_test)
```
